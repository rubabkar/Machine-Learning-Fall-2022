## The following code is for transforming images

```python
def convertImage(image, new_name):
    img = Image.open(image)
    img = img.convert("RGBA")
    
    datas = img.getdata()
    
    newData = []
    
    for item in datas:
        if item[0] >= 150 and item[1] >= 150 and item[2] >= 150:
            newData.append((0, 0, 0, 0))
        else:
            newData.append(item)
            
    img.putdata(newData)
    img.save(new_name, "PNG")
    print("Successfully done")
```

```python
convertImage('data/data1/image-cats/cat2.png', 'data/Modded_cat2.png')
```
